```python
import random
def user_input():
    print("you have choices 1. snake 2. water 3. gun")
    choice = input("Enter your choice: ")
    while choice not in ['snake', 'water', 'gun']:
        print("Invalid choice, Please enter again.....")
        choice = input("Enter your choice again: ")
    return choice
def computer_input():
    return random.choice(['snake', 'water', 'gun'])
def detetmine_winner(user, computer):
    if user == computer:
        return "draw!!"
    elif((user == 'snake' and computer == 'water')or
            (user=='water' and computer == 'gun') or 
            (user == 'gun' and computer == 'snake')):
        return "you win!!"
    else:
        return "computer's win!!"
def play():
    print("welcome to the game of Snake, Water, Gun!")
    user_score = 0
    computer_score = 0
    rounds = int(input("how many rounds do you want to play? "))
    for i in range(1,rounds+1):
        print(f"\n---round {i}---\n")
        computer = computer_input()
        user = user_input()

        print(f"your choice is {user}")
        print(f"computer's choice is {computer}")

        result = detetmine_winner(user, computer)
        if(result == "draw!!"):
            print("It's a draw!")
        elif(result == "you win!!"):
            print("You win this round!")
            user_score += 1 
        else:
            print("Computer wins this round!")
            computer_score += 1
    print("\n---Final Scores---")
    print(f"score -> You: {user_score} | Computer: {computer_score}")
    if user_score > computer_score:
        print("Congratulations! You are winner!")
    elif user_score < computer_score:
        print("Computer is the winner!")
    else:
        print("It's draw!")
if __name__ == "__main__":
    play()
```

```python
import pandas as pd
data= {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [24, 30, 22, 35, 28],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'Sales'],
    'Salary': [50000, 60000, 55000, 70000, 65000]

}

df = pd.DataFrame(data)
df
```

iloc in Pandas -> Access by integer based index positions



```python
#Row at index 0
df.iloc[0]
```

```python
#First three rows
df.iloc[:3]
```

```python
#last three rows
df.iloc[-3:]
```

```python
# Row 1, Column 3 
df.iloc[1,3]
```

```python
# Rows 1 to 3, Cloumns 1 to 2
df.iloc[1:4, 1:3]
```

```python
#last row
df.iloc[-1]
```

loc in pandas -> Access by labels(row index, cloumn names)

```python
#get row at index 2
df.loc[2]
```

```python
# All rows, only name and Salary columns
df.loc[:, ['Name', 'Department', 'Salary']]
```

```python
#Filter rows where Salary is greater than 60000
df.loc[df['Salary']>60000]
```

```python
#Update Salary for Bob
df.loc[1, 'Salary'] = 100000
df
```

```python
# Get names of all employees selary greater than 60000.
df.loc[df['Salary']>60000, 'Name']
```

```python
# Retrieve the salary of the 2nd employee using iloc.
df.iloc[1,3]
```

```python
# Use .loc to get the department of Eva.
df.loc[df['Name']=='Eva', 'Department']
```

```python
# Slice the DataFrame to get rows 0 to 2 and all columns using iloc
df.iloc[0:3, :]
```

```python
# We can also change row lables using the index attribute.
df.index = ['A', 'B', 'C', 'D', 'E']
df
```
