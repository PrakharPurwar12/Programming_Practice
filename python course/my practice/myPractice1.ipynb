Data Visualization


```python
plt.figure(figsize = (6, 6))
plt.scatter(df['Duration'], df['Calories'])
plt.title('Duration vs Calories')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.show()
```

```python
plt.hist(df['Calories'], bins=10, edgecolor='black')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.bar(df['Duration'], df['Calories'], color='red')
plt.title('Calories Burned by Duration')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.grid()
plt.show()
```
