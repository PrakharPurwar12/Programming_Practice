```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('churn.csv')  
df.head(10)
```

```python
print(df.shape)
print(df.dtypes)
```

```python
# Check for missing values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Basic statistics
print("\nStatistics:\n", df.describe())
```

### Target Variable (Churn) Distribution

```python
# Countplot for target variable
sns.set(style="whitegrid")
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df, palette='Set2')
plt.title("Customer Churn Distribution")
plt.xticks([0, 1], ["Stayed (0)", "Churned (1)"])
plt.ylabel("Number of Customers")
plt.xlabel("Churn Status")
plt.show()
```

# Categorical Feature Analysis

```python
plt.figure(figsize=(5, 4))
sns.countplot(x='Gender', hue='Churn', data=df, palette='pastel')
plt.title("Gender vs Churn")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Subscription Type', hue='Churn', data=df, palette='muted')
plt.title("Subscription Type vs Churn")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Contract Length', hue='Churn', data=df, palette='coolwarm')
plt.title("Contract Length vs Churn")
plt.show()
```

# Numerical Feature Distribution

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='Age', bins=20, kde=True, color='skyblue')
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Churn', y='Tenure', data=df, palette='Set2')
plt.title("Tenure vs Churn")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['Total Spend'], kde=True, bins=20, color='green')
plt.title("Total Spend Distribution")
plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(10, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# Feature vs Churn (for insights)

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='Churn', y='Support Calls', data=df)
plt.title("Support Calls vs Churn")
plt.show()
```
