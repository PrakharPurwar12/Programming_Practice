### Slicing

```python

nums = [10, 20, 30, 40, 50, 60]

print(nums[1:4])   
print(nums[:3])    
print(nums[-3:])   
print(nums[::-1])  
```

```python
print(nums[-4:-1])
```

```python
print(nums[-2:-4])
```

### Sample DataFrame

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Department': ['HR', 'Finance', 'Engineering', 'Marketing', 'Sales'],
    'Salary': [50000, 60000, 70000, 80000, 90000]
}

df = pd.DataFrame(data)
df
```

 ### .iloc[] in Pandas
*** Access by integer-based index positions ***

```python
# Row at index 0
#df.iloc[0]

# First 3 rows
#df.iloc[:3]

# Row 1, column 2 (Age of Bob)
#df.iloc[1, 1]

# Rows 1 to 3 and columns 1 to 2
 #df.iloc[1:4, 1:3]

# Last row
#df.iloc[-1]
```

### .loc[] in Pandas
Access by labels (row index, column names).

```python
# Get row at index 2 (Charlie)
#df.loc[2]

# All rows, only Name and Salary
#df.loc[:, ['Name', 'Salary']]

# Rows 1 to 3 (inclusive) for Name and Department
#df.loc[1:3, ['Name', 'Department']]

# Filter rows where Salary > 60000
#df.loc[df['Salary'] > 60000]

# Update Age for Bob
#df.loc[1, 'Age'] = 31
```

### Practice Problems

1. Get names of all employees older than 30.

2. Retrieve the salary of the 2nd employee using iloc.

3. Use .loc to get the department of Eva.

4. Slice the DataFrame to get rows 0 to 2 and all columns using iloc.

```python
# 1 Get names of all employees older than 30.
df.loc[df['Age'] > 30, 'Name']
```

```python
# Retrieve the salary of the 2nd employee using iloc.
df.iloc[1, 3]
```

```python
#Use .loc to get the department of Eva.
df.loc[df['Name'] == 'Eva', 'Department'].values[0]
```

```python
#Slice the DataFrame to get rows 0 to 2 and all columns using iloc.
df.iloc[0:3, :]
```
