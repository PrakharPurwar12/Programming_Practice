# **Vectorization & Broadcasting**


```python
# Element-wise operations (NO LOOPS!)
arr = np.array([[1, 2, 3], [4,3,6]])
squares = arr ** 2             
print(squares)
```

```python
a = np.array([1, 2, 3])

# Add scalar to array
result = a + 5
print(result)  # Output: [6 7 8]
#Explanation: 5 is broadcast to [5, 5, 5] internally.
```

```python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

# Vectorized addition
result = a + b
print(result)  
```

```python
#Broadcasting works across dimensions.
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

row = np.array([10, 20, 30])

# Broadcast row to each row of matrix
result = matrix + row
print(result)
#Explanation: The 1D row is "stretched" across each row of the matrix.
```

```python
col = np.array([[1],
                [2],
                [3]])  # 3*1
row = np.array([10, 20, 30])  # 1*3

result = col + row
print(result)
#Explanation: Column (3×1) and row (1×3) are broadcast into a (3×3) matrix.
```

```python
#Element-wise Multiplication of Different Shapes
a = np.array([[1, 2, 3],
              [4, 5, 6]])

scaling_factors = np.array([1, 10, 100])  # shape (3,)

# Each column gets multiplied by its corresponding factor
result = a * scaling_factors
print(result)
```

# **Mathematical Operations**

```python
# Aggregations
data = np.random.normal(0, 1, 1000)  # Normal distribution
print(f"Mean: {data.mean():.2f}, STD: {data.std():.2f}")
```

```python
# Linear algebra
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
dot_product = A @ B  # Matrix multiplication
determinant = np.linalg.det(A)
print(dot_product)
print(determinant)
```

### comparing numpy operations with normal operations

```python
import random
import time

# Create sample marks for 10000 students (5 subjects each)
students = [[random.randint(40, 100) for _ in range(5)] for _ in range(10000)]

start = time.perf_counter_ns()
# Compute total and average using loops
totals = []
averages = []
for student in students:
    total = sum(student)
    avg = total / len(student)
    totals.append(total)
    averages.append(avg)
    
end = time.perf_counter_ns()
print("Time taken (Python list):", round(end - start), "nanoseconds")
```

```python
import numpy as np
import time

# Convert the same list to a NumPy array
data = np.array(students)

start = time.perf_counter_ns()

# Vectorized computation
totals_np = data.sum(axis=1)
averages_np = data.mean(axis=1)

end = time.perf_counter_ns()
print("Time taken (NumPy):", round(end - start), "nanoseconds")
```
