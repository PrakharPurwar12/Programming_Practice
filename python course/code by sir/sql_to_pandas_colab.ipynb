# 🧠 SQL to Pandas: A Hands-on Guide for AI/ML Students
This notebook introduces SQL basics and shows how to achieve the same functionality using Pandas.
We'll work with a sample `students.db` SQLite database.

````python
# 📦 Setup: Load packages and create a SQLite database
import sqlite3
import pandas as pd
import numpy as np

# Create a connection
conn = sqlite3.connect('students.db')

# Create a sample table
conn.execute('''CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    math_score INTEGER,
    english_score INTEGER,
    grade TEXT
);''')

# Insert sample data
conn.execute("DELETE FROM students;")
students_data = [
    (1, 'Alice', 85, 78, 'A'),
    (2, 'Bob', 67, 88, 'B'),
    (3, 'Charlie', 92, 95, 'A'),
    (4, 'David', 58, 62, 'C'),
    (5, 'Eva', 76, 83, 'B')
]
conn.executemany("INSERT INTO students VALUES (?, ?, ?, ?, ?);", students_data)
conn.commit()
````

## 🔍 SQL Queries: Examples
Let’s run some SQL queries and then replicate the same logic using Pandas.

````python
# Run SQL query: Select all students
query = "SELECT * FROM students;"
df_sql = pd.read_sql_query(query, conn)
df_sql
````

## 🔁 Pandas Equivalent of SQL Queries
Let’s load the same data into Pandas and perform equivalent operations.

````python
# Load data into Pandas
df = pd.read_sql_query("SELECT * FROM students", conn)
df.head()
````

### 🎯 Example 1: Students scoring above 80 in Math (SQL vs Pandas)

````python
# SQL Equivalent
pd.read_sql_query("SELECT * FROM students WHERE math_score > 80", conn)
````

````python
# Pandas Equivalent
df[df['math_score'] > 80]
````

### 📊 Example 2: Group by grade and count students

````python
# SQL Equivalent
pd.read_sql_query("SELECT grade, COUNT(*) as student_count FROM students GROUP BY grade", conn)
````

````python
# Pandas Equivalent
df.groupby('grade').size().reset_index(name='student_count')
````
