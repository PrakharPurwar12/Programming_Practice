### Iterating and Aggregations

```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Marks': [95, 85, 92, 88, 76],
    'Grade': ['A', 'B', 'A', 'B', 'C']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
#  Aggregation using groupby()
avg_marks_by_grade = df.groupby('Grade')['Marks'].mean()
print("\nAverage Marks by Grade:")
print(avg_marks_by_grade)
```

### Converting Data
 #### To Dictionary / NumPy / List


```python
import pandas as pd

# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Marks': [95, 85, 92],
    'Grade': ['A', 'B', 'A']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# ✅ Convert to Dictionary
dict_data = df.to_dict()
print("\nConverted to Dictionary (default - columns):")
print(dict_data)

# ✅ Convert to NumPy Array
numpy_data = df.to_numpy()
print("\nConverted to NumPy Array:")
print(numpy_data)
```
